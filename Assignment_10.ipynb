```python
import pandas as pd
link = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv"
data = pd.read_csv(link)
df = pd.DataFrame(data)
df.head(5)
```

```python
# 1. Delete unnamed columns
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.head(5)
```

```python
# 2. Show the distribution of male and female
pd.value_counts(df['Gender'])
```

```python
# 3.Show the top 5 most preferred names

names = df.groupby('Name').Count.agg(['sum']) # Distinct names with counts
df_names = pd.DataFrame(names)
df_names.sort_values(by='sum', ascending=False, inplace=True) 
# sorting names in descending order of count, in order to get the most preferred names 

df_names.head(5) # top 5 names of most preferred names
```

```python
# 4. What is the median name occurence in the dataset
import math # import math library
dist_count = df_names.count() # Get total number of names
name_med = math.ceil(((dist_count+1)/2)) # calculate median value, formula = (n+1)/2
df.loc[name_med,:] # details of median name
```

```python
# 5.Distribution of male and female born count by states
# a) Male names born by states
male_names = (df.loc[df['Gender'].isin(['M'])]).groupby('State').Count.agg(['sum'])
male_names.sort_values(by='sum', ascending=True, inplace=True)
male_names
```

```python
# 5.Distribution of male and female born count by states
# a) Female names born by states
fm_names = (df.loc[df['Gender'].isin(['F'])]).groupby('State').Count.agg(['sum'])
fm_names.sort_values(by='sum', ascending=True, inplace=True)
fm_names
```
